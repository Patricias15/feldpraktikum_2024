```python
import numpy as np
import pandas as pd 
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
out = pd.read_csv('0406_Turnau_outside.csv',skiprows=16)
inside = pd.read_csv('0406_Turnau_inside.csv',skiprows=16)
```

```python
time_out = pd.to_datetime(out['Time'].values)
time_in = pd.to_datetime(inside['Time'].values)
```

```python
fig , ax = plt.subplots(1, figsize=(20,10))
ax.plot(time_out, out['Concentration'], color='blue', label= '$CPC_{outside}$')
ax.plot(time_in, inside['Concentration'], color= 'red',label= '$CPC_{inside}$')
ax.set_yscale('log')
ax.xaxis.set_minor_locator(mdates.MinuteLocator(byminute=[0,15,30,45,60]))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M:%S"))
ax.set_title('CPC Comparison', fontsize= 20)
ax.set_ylabel('Particle concentration #/ $cm^3$', fontsize= 20)
ax.set_xlabel('Time UTC', fontsize= 20)
ax.tick_params(axis='both', which='minor', labelsize=40)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylim(1e2,1000000)
plt.grid()

plt.legend(loc='upper left', fontsize=25)
plt.savefig('Comparison.png')
fig.show()
```
